Here's the boilerplate (starter code) for a class called Faction that you'll be working on. The instances of this class will represent the different competing factions.

```python
# imports here

import random
import time
import matplotlib.pyplot as plt


Fruits = ("apples", "oranges", "bananas", "strawberries")
maximum_amount = 200

class Faction:
    """
    Represents a faction in the fruit wars.
    Each faction has a name and controls 4 types of fruits.
    Think of it like: Faction = Team + Inventory of fruits
    """

    def __init__(self, name: str, apples: int, oranges: int, bananas: int, strawberries: int) -> None:
        """
        Initialize a faction with a name and fruit amounts.

        """
        # Use setters to validate everything
        self.name = name
        self.apples = apples
        self.oranges = oranges
        self.bananas = bananas
        self.strawberries = strawberries

    # ===== GETTERS AND SETTERS =====
    
    @property
    def name(self) -> str:
        """Get the faction's name"""
        return self._name
    
    @name.setter
    def name(self, name: str) -> None:
        """Set and validate the faction's name"""
        if not isinstance(name, str):
            raise TypeError("Faction name must be a text (a string)")
        name = name.strip()
        if not name:
            raise ValueError("Faction name cannot be empty")
        self._name = name

    @property
    def apples(self) -> int:
        """Get the number of apples"""
        return self._apples
    
    @apples.setter
    def apples(self, apples: int) -> None:
        """Set and validate apple amount"""
        if not isinstance(apples, int):
            raise TypeError("Apples amount should be a number")
        if apples < 0 or apples > maximum_amount:
            raise ValueError(f"Apples amount must be between 0 and {maximum_amount}")
        self._apples = apples

    @property
    def oranges(self) -> int:
        """Get the number of oranges"""
        return self._oranges
    
    @oranges.setter
    def oranges(self, oranges: int) -> None:
        """Set and validate orange amount"""
        if not isinstance(oranges, int):
            raise TypeError("Oranges amount should be a number")
        if oranges < 0 or oranges > maximum_amount:  #  checking 'apples'
            raise ValueError(f"Oranges amount must be between 0 and {maximum_amount}")
        self._oranges = oranges

    @property
    def bananas(self) -> int:
        """Get the number of bananas"""
        return self._bananas
    
    @bananas.setter
    def bananas(self, bananas: int) -> None:
        """Set and validate banana amount"""
        if not isinstance(bananas, int):
            raise TypeError("Bananas amount should be a number")
        if bananas < 0 or bananas > maximum_amount:
            raise ValueError(f"Bananas amount must be between 0 and {maximum_amount}")
        self._bananas = bananas

    @property
    def strawberries(self) -> int:
        """Get the number of strawberries"""
        return self._strawberries
    
    @strawberries.setter
    def strawberries(self, strawberries: int) -> None:
        """Set and validate strawberry amount"""
        if not isinstance(strawberries, int):
            raise TypeError("Strawberries amount should be a number")
        if strawberries < 0 or strawberries > maximum_amount:
            raise ValueError(f"Strawberries amount must be between 0 and {maximum_amount}")
        self._strawberries = strawberries 


  
    
    
        

    # ===== REQUIRED METHODS (TO BE IMPLEMENTED) =====

    def inventory(self) -> int:

        """ Return total amount of all fruits"""
        return self.apples+self.oranges+self.bananas+self.strawberries
      
        

    def status(self) -> None:

        """ Print faction's status based on total fruits"""
        total = self.inventory()
        print(f"----Status for {self.name} (total fruits :{total})-----")
        if total <= 150:
            print(f"This aint nothing but a scratch! {self.name} might be struggling for the moment, but will be back with vengance!")
        elif 150 < total <= 500:
            print(f"The proud {self.name} is very eager to toss some fruits! Where are the opponents?!")
        else:
            print(f"{self.name} is GODLIKE! Their position is super dominant!")    
        

        

    def visualize_resources(self) -> None:
        """Show a bar chart of fruits (color-coded)"""

        labels = list(Fruits)
        values = [self.apples, self.oranges, self.bananas, self.strawberries]
        colors = []

        # for v in values:
        #     if v >= 100:
        #         colors.append("green")
        #     elif 50 < v < 100:
        #         colors.append("orange")
        #     else:
        #         colors.append("red")
        colors = ["green" if v >= 100 else "orange" if v >
                  50 else "red" for v in values]


        plt.figure(figsize = (8,5))
        bars = plt.bar(labels, values, color = colors, edgecolor = "black")
        plt.ylim(0, maximum_amount + 10)
        plt.title(f"{self.name} - Resource inventory")
        plt.xlabel("Fruit")
        plt.ylabel("Amount")
        #Adding numbers on bars:
        for bar, value in zip(bars, values):
            plt.text(bar.get_x() + bar.get_width() / 2, value + 2, str(value), ha="center", va="bottom", fontsize=9)
        plt.tight_layout()  # adjusts padding so text doesnâ€™t overlap
        plt.show()



    

    def organical_growth(self) -> None:

        """ Randomly grow fruits (1-5 per fruit) unless already max(200)"""

        growths = {
            "apples" : random.randint(1,5),
            "oranges": random.randint(1,5),
            "bananas": random.randint(1,5),
            "strawberries":random.randint(1,5)
        }

        for fruit_name, grow in growths.items():
            current = getattr(self, fruit_name)
            if current >= maximum_amount:
                print(f"Glorious {self.name} is already at maximum capacity for {fruit_name}")
                continue
            new_amount =  min(maximum_amount, current+grow)
            actual_growth = new_amount - current
            setattr(self, fruit_name, new_amount)
            print(f"Glorious {self.name} has harvested {actual_growth} {fruit_name}, to a new total of {new_amount}")

    def __mul__(self, other: 'Faction') -> None:

        """Attack another faction (use * operator)
        - 70% success chance to steal.
        - Base steal is 20% of each fruit (rounded down).
        - Independent 10% chance to be CRITICAL and steal 40% instead.
        - If defender has zero total resources at start, attacker must pick another target (we just print).
        - Each fruit can't exceed maximum_amount for attacker; any overflow becomes 'casualties' (lost).
        - If defender has <= 5 of a fruit, attacker steals all of that fruit (humiliating blow)."""
        if not isinstance(other, Faction):
            raise TypeError("Can only attack another Faction")
        if self is other:
            print(f"{self.name} cannot attack itself")
            return

    # If defender has nothing, no attack
        if other.inventory() == 0:
            print(
                f"{other.name} has no resources left. No glory in attacking the weak!")
            return
        print(f"\n{self.name} is going to attack on {other.name}!")
        time.sleep(0.6)

    # 70% chance the attack succeeds
        roll = random.random()  # random float between 0 and 1
        success = roll < 0.7
        print(f"Attack roll: {roll:.2f} -> {'SUCCESS' if success else 'FAILURE'}")
        time.sleep(0.8)

        if not success:
            print(f"{other.name} defended bravely. Attack failed!")
            return

    # Critical chance (10%)
        crit_roll = random.random()
        critical = crit_roll < 0.10
        percentage = 0.4 if critical else 0.2
        print(
            f"Critical roll: {crit_roll:.3f} -> {'CRITICAL HIT! (+40%)' if critical else 'Normal hit (+20%)'}")
        time.sleep(0.8)

        # Steal fruit from defender
        for fruit in Fruits:
            current_def = getattr(other, fruit)
            current_att = getattr(self, fruit)

            if current_def <= 5:
                stolen = current_def
            else:
                stolen = int(current_def * percentage)
            

            # Control max limit (200 )
            space_left = maximum_amount - current_att
            actual_taken = min(stolen, space_left)

        # Update attacker and defender
            setattr(self, fruit, current_att + actual_taken)
            setattr(other, fruit, current_def - actual_taken)

            print(f"{self.name} stole {actual_taken} {fruit} from {other.name}.")

        print(
            f"\nAttack complete! {self.name} now has {self.inventory()} fruits.")
        print(f"{other.name} is reduced to {other.inventory()} fruits...")


    def trade_offer(self, other: 'Faction', offered_fruit_name: str, offered_amount: int, 
                    requested_fruit_name: str, requested_amount: int) -> None:
        """ Attempt to trade resources between two factions.
        This method allows the current faction to offer a fruit in exchange for another fruit
         from a target faction. A trade can only be completed if:
      - Both factions own enough of the respective fruits
      - The trade does not push any fruit above the maximum storage limit
      - The offer is fair (offered_amount >= requested_amount)
        If the offer is unfair (offered < requested), the target faction becomes insulted and
        immediately launches an attack instead of trading.
        When all conditions are met, the trade goes through and both sides exchange resources.
        A summary of the result is printed to the terminal."""
        
        if self is other:
            print("A faction cannot trade with itself.")
            return
        
        if offered_fruit_name not in Fruits or requested_fruit_name not in Fruits:
            print("Invalid fruit name in trade. Trade cancelled.")
            return
        
        if getattr(self, offered_fruit_name) < offered_amount : 
            print("Trade failed because you do not have enough for trading")
            return
        if getattr(other, requested_fruit_name) < requested_amount:
            print("Trade can not be executed as they do not have what you want")
            return
        
        if offered_amount < requested_amount :
            print(f"{other.name} is insulted by this offer and  wan to make an Attack as revenge! ")
            other * self
            return
        if getattr(self, requested_fruit_name) + requested_amount > maximum_amount:
            print("Trade failed as you would exceed fruit limit(200)") 
            return
        if getattr(other, offered_fruit_name) + offered_amount > maximum_amount:
            print("Trade failed as other part would exceed fruit limit(200)")
            return
        
        #Now we will set the trade:
        setattr(self, offered_fruit_name, getattr(self, offered_fruit_name) - offered_amount)
        setattr(self, requested_fruit_name, getattr(self, requested_fruit_name) + requested_amount)
        setattr(other, offered_fruit_name , getattr(other, offered_fruit_name) +offered_amount )
        setattr(other, requested_fruit_name, getattr(other, requested_fruit_name)-requested_amount)

        print(f"Trade succeed !!: {self.name} traded {offered_amount} {offered_fruit_name} for {requested_amount} {requested_fruit_name } with {other.name}")
        print(f"{self.name} now has {self.inventory()} fruits total")
        print(f"{other.name} now has {other.inventory()} fruits total")
        
        

    def __repr__(self) -> str:
        """ String representation of faction"""
        return (f"Faction(name={self.name!r}, apples={self.apples}, oranges={self.oranges}, bananas={self.bananas}, strawberries={self.strawberries})")
        
```
